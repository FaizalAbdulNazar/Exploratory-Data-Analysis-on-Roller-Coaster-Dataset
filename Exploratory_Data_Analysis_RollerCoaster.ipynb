```python
df = df.rename(columns = {'coaster_name': 'Coaster_Name', 'year_introduced': 'Year_Introduced',
                     'opening_date_clean':'Opening_Date',
                     'speed_mph':'Speed_mph','height_ft':'Height_ft',
                     'Inversions_clean':'Inversions','Gforce_clean':'Gforce' })
```

```python
#Checking if there is any duplicates in a particular row
df.loc[df.duplicated(subset = ['Coaster_Name'])]
```

```python
#Checking a duplicate
df.query('Coaster_Name == "Derby Racer"')
```

```python
#Checking duplicates and taking the inverse('~') of them. The need of inverse is to identify the original data and to remove all the duplicates.
#After taking out the original rows, the index should be modified.

df = df.loc[~df.duplicated(subset = ['Coaster_Name', 'Location', 'Opening_Date'])] \
.reset_index(drop = True).copy()
```
